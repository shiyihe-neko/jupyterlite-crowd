```python
import requests
import json
import os

def upload_results():
    """
    ç®€å•ä¸Šä¼ å‡½æ•°ï¼šä¸Šä¼ æ¸…ç†åçš„ CSV å’Œå½“å‰ notebook åˆ°æœ¬åœ°åç«¯ã€‚
    """
    participant_id = "testuser"
    base_url = "http://127.0.0.1:5050"   # ä½ çš„ Flask åç«¯åœ°å€

    try:
        # === ä¸Šä¼ æ¸…ç†åçš„ CSV ===
        csv_path = "/drive/datasets/demo_cleaned.csv"
        if not os.path.exists(csv_path):
            raise FileNotFoundError(f"{csv_path} ä¸å­˜åœ¨ï¼Œè¯·å…ˆä¿å­˜æ¸…ç†åçš„æ–‡ä»¶ã€‚")

        with open(csv_path, "rb") as f:
            res1 = requests.post(
                f"{base_url}/upload_file",
                params={"participant_id": participant_id},
                files={"file": f}
            )
        print(f"âœ… CSV ä¸Šä¼ æˆåŠŸ: {res1.status_code}")
        print(res1.text)

        # === ä¸Šä¼ å½“å‰ notebook ===
        nb_path = "task.ipynb"
        if not os.path.exists(nb_path):
            raise FileNotFoundError(f"{nb_path} ä¸å­˜åœ¨ï¼Œè¯·ç¡®è®¤ notebook åç§°ã€‚")

        with open(nb_path, "rb") as f:
            res2 = requests.post(
                f"{base_url}/upload_notebook",
                params={"participant_id": participant_id},
                files={"file": f}
            )
        print(f"âœ… Notebook ä¸Šä¼ æˆåŠŸ: {res2.status_code}")
        print(res2.text)

    except Exception as e:
        print(f"âŒ ä¸Šä¼ å¤±è´¥ï¼š{e}")

upload_results()
```

```python
import requests, os

def upload_to_backend(file_path, participant_id, kind="data"):
    """ä¸Šä¼ æ–‡ä»¶åˆ°åç«¯ï¼ˆdata=CSV, notebook=ipynbï¼‰"""
    url = {
        "data": "http://192.168.0.111:5050/upload_file",
        "notebook": "http://192.168.0.111:5050/upload_notebook",
    }[kind]

    with open(file_path, "rb") as f:
        res = requests.post(url, files={"file": f}, data={"participant_id": participant_id})
    print(f"âœ… ä¸Šä¼  {kind} æˆåŠŸ: {res.status_code}, {res.json()['path']}")
    return res.json()

# ä½¿ç”¨ç¤ºä¾‹
upload_to_backend("demo_cleaned_111.csv", "shiyi", kind="data")
upload_to_backend("task_2.ipynb", "shiyi", kind="notebook")
```
